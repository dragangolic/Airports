```python
# Split the 'Data' column into separate columns
df2[['Airport', 'Location', 'Area']] = df2['Data'].str.split(' — ', expand=True)
df2
```

```python
# Drop the 'Data' column
df2.drop('Data', axis=1, inplace=True)
df2
```

```python
# Deleting empty rows 
rows_to_delete = [1, 3, 5, 7, 9, 11, 13, 15, 17]
df2.drop(rows_to_delete, inplace=True)
df2.reset_index(drop=True, inplace=True)
df2
```

```python
# Split the 'Location' column into 'City' and 'Country' columns
df2[['City', 'Country']] = df2['Location'].str.split(' ', n=1, expand=True)
df2
```

```python
# Split the 'Area' column into 'square km' and 'square miles' columns
df2[['Square km', 'Square mi']] = df2['Area'].str.split(' ', n=1, expand=True)
df2
```

```python
columns_to_delete = ['Location','Area']
df2.drop(columns=columns_to_delete, inplace=True)
df2
```

```python
df2['Square mi'] = df2['Square mi'].str.replace('km2', '').str.replace('(mi2)', '').str.replace('(', '').str.replace(')', '')
df2
```
